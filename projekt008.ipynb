```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

S – osobniki podatne na zachorowanie (ang. susceptible).

I – osobniki zakażone (ang. infectious).

R – osobniki odporne (ang. recovered lub removed), które już nigdy nie zachorują.

$$ \frac{d S}{dt} = - \beta SI $$

$$ \frac{d S}{dt} = - \beta SI $$
$$ \frac{d I}{dt} = \beta SI - \gamma I $$
$$ \frac{d R}{dt} = \gamma I $$

 $\beta$ – prawdopodobieństwo przekazania w jednostce czasu (np. w ciągu doby) choroby każdemu z osobników, z którym zarażony ma kontakt, $\beta$ $\epsilon$ (0, 1)
 i im większą przyjmuje wartość, tym bardziej zaraźliwa jest choroba.

 $\gamma$ – prawdopodobieństwo wyzdrowienia w jednostce czasu, $\gamma$ $\epsilon$ (0, 1)
 i im większą przyjmuje wartość, tym szybciej następuje wyzdrowienie.

```python
beta = 0.5
gamma = 0.5

def func(y, t):
    S, I, R = y
    return -beta*S*I, beta*S*I - gamma*I, gamma*I

S0 = 9
I0 = 1
R0 = 0

ts1 = np.linspace(0, 10, 1000)
res1 = odeint(func, (S0, I0, R0), ts1)
```

```python
plt.plot(ts1, res1[:, 0], label="S")
plt.plot(ts1, res1[:, 1], label="I")
plt.plot(ts1, res1[:, 2], label="R")

plt.legend()
```
