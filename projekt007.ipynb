```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
# bez tłumienia i wymuszenia

x = sp.Function('x')
w0 = sp.Symbol('omega_0', real = True, positive = True)
t = sp.Symbol('t')
x0 = sp.Symbol('x_0', real = True) # początkowe położenie
v0 = sp.Symbol('v_0', real = True) # początkowa prędkość

x(t), w0, t, x0, v0
```

```python
eq = sp.Eq(x(t).diff(t, t), -w0**2 * x(t))
eq
# równanie oscylatora harmonicznego bez tłumienia i wymuszenia
```

```python
sol = sp.dsolve(eq, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
# ics - warunki brzegowe
```

```python
sol.lhs, sol.rhs
```

```python
sol.rhs.subs({x0: 1, v0: -1, w0: 2})

# podstawianie wartości
```

```python
sp.plot(sol.rhs.subs({x0: 1, v0: -1, w0: 2}), (t, -5, 5))
```

```python
solf = sp.lambdify(t, sol.rhs.subs({x0: 1, v0: -1, w0: 2}))
# zmiana na funkcję numpy'ową
ts = np.linspace(-5, 5, 1000)
plt.plot(ts, solf(ts))
plt.grid()
```

```python
x = sp.Function('x')
w0 = sp.Symbol('omega_0', real = True, positive = True)
t = sp.Symbol('t')
x0 = sp.Symbol('x_0', real = True) # początkowe położenie
v0 = sp.Symbol('v_0', real = True) # początkowa prędkość
```

```python
# z tłumieniem

beta = sp.Symbol('beta') # współczynnik tłumienia (ten parametr zmieniamy)

eq1 = sp.Eq(x(t).diff(t, t) + 2 * beta * x(t).diff(t) + w0**2 * x(t), 0)
eq1
# równanie oscylatora harmonicznego
```

```python
sol1 = sp.dsolve(eq1, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol1
```

```python
sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 1})

# podstawianie wartości
```

```python
sp.plot(sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 1}), (t, -5, 5))
```

```python
solf13 = sp.lambdify(t, sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 1.5}))
# zmiana na funkcję numpy'ową
ts1 = np.linspace(-5, 5, 1000)
plt.plot(ts1, solf13(ts1))
plt.grid()
```

```python
solf1 = sp.lambdify(t, sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 1}))
# zmiana na funkcję numpy'ową
ts1 = np.linspace(-5, 5, 1000)
plt.plot(ts1, solf1(ts1))
plt.grid()
```

```python
solf12 = sp.lambdify(t, sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 0.5}))
# zmiana na funkcję numpy'ową
ts1 = np.linspace(-5, 5, 1000)
plt.plot(ts1, solf12(ts1))
plt.grid()
```

```python
solf14 = sp.lambdify(t, sol1.rhs.subs({x0: 1, v0: -1, w0: 2, beta: 0.1}))
# zmiana na funkcję numpy'ową
ts1 = np.linspace(-5, 5, 1000)
plt.plot(ts1, solf14(ts1))
plt.grid()
```

```python
# z wymuszeniem bez tłumienia

A = sp.Symbol('A') # amplituda przyspieszenia wymuszającego
w = sp.Symbol('omega') # częstość siły wymuszającej (to właśnie ten parametr zmieniamy)

eq2 = sp.Eq(x(t).diff(t, t) + w0**2 * x(t), A * sp.cos(w * t))
eq2
# równanie oscylatora harmonicznego
```

```python
sol2 = sp.dsolve(eq2, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol2
```

```python
sol2.rhs.subs({x0: 1, v0: -1, w0: 2, A: 1, w: 1})

# podstawianie wartości
```

```python
sp.plot(sol2.rhs.subs({x0: 1, v0: -1, w0: 2, A: 1, w: 1}), (t, -5, 5))
```
